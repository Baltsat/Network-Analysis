## ***Задание №3. Средняя длина пути***

***Так как графы не являются сильно связными, придётся вычислять среднюю длину пути для каждой отдельной вершины, после чего считать среднее расстояние.***

```python
both_list = []
fbl_list = []
bfsistema_list = []
for A in (G1.subgraph(a).copy() for a in nx.connected_components(G1)):
    both_list.append(nx.average_shortest_path_length(A))
for B in (G2_1.subgraph(b).copy() for b in nx.connected_components(G2_1)):
    fbl_list.append(nx.average_shortest_path_length(B))
for C in (G3_1.subgraph(c).copy() for c in nx.connected_components(G3_1)):
    bfsistema_list.append(nx.average_shortest_path_length(C))
print(f"Средняя длина пути для пересечения обоих сообществ равна {mean(both_list)}")
print(f"Средняя длина пути для сообщества 'fbl.russia' равна {mean(fbl_list)}")
print(f"Средняя длина пути для сообщества 'bfsistema' равна {mean(bfsistema_list)}")
```

## ***Задание №4. Какая модель социальной сети?***

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()

def l_average(graph):
  largest_cc = max(nx.connected_components(graph), key=len)
  return nx.average_shortest_path_length(graph.subgraph(largest_cc).copy())
```

```python
print('Количество вершин: {}'.format(G2_1.number_of_nodes()))
print('Количество рёбер: {}'.format(G2_1.number_of_edges()))
print('Среднее количество соседей у узлов в графе: {}'.format(round(G2_1.number_of_edges() / float(G2_1.number_of_nodes()), 4)))
plot_degree_dist(G2_1)

print('Коэффициент кластеризации: {}'.format(nx.average_clustering(G2_1)))
print('<L>: {}'.format(l_average(G2_1)))
```

```python
def calculate_metrics(num_vertices, num_edges):
    # Random graph
    random_graph = nx.gnm_random_graph(num_vertices, num_edges)
    random_c = nx.average_clustering(random_graph)
    random_l = l_average(random_graph)

    # BA (Barabasi-Albert) model
    ba_graph = nx.barabasi_albert_graph(num_vertices, int(num_edges/num_vertices))
    ba_c = nx.average_clustering(ba_graph)
    ba_l = l_average(ba_graph)

    # WS (Watts-Strogatz) model
    # ws_graph = nx.watts_strogatz_graph(num_vertices, int(num_edges/num_vertices), 0.5)
    # ws_c = nx.average_clustering(ws_graph)
    # ws_l = l_average(ws_graph)

    return random_c, random_l, ba_c, ba_l#, ws_c, ws_l

largest_cc = max(nx.connected_components(G3_1), key=len)
largest_cc = G3_1.subgraph(largest_cc).copy()


print(largest_cc.number_of_edges(), largest_cc.number_of_nodes())

num_vertices = largest_cc.number_of_nodes()
num_edges = largest_cc.number_of_edges()
random_c, random_l, ba_c, ba_l = calculate_metrics(num_vertices, num_edges)

print("Random graph:")
print("Clustering coefficient (C):", random_c)
print("Average path length (<L>):", random_l)
print()
print("BA (Barabasi-Albert) model:")
print("Clustering coefficient (C):", ba_c)
print("Average path length (<L>):", ba_l)
print()
# print("WS (Watts-Strogatz) model:")
# print("Clustering coefficient (C):", ws_c)
# print("Average path length (<L>):", ws_l)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

N = 117

# Random Network
random_graph = nx.erdos_renyi_graph(n=N, p=0.1)  # Example: 1000 nodes, edge probability of 0.1
random_degrees = [deg for _, deg in random_graph.degree()]
random_degree_counts = np.bincount(random_degrees)
random_x = np.arange(len(random_degree_counts))
random_y = random_degree_counts / float(sum(random_degree_counts))

# BA (Barabasi-Albert) Model
ba_graph = nx.barabasi_albert_graph(n=N, m=5)  # Example: 1000 nodes, each new node connects to 5 existing nodes
ba_degrees = [deg for _, deg in ba_graph.degree()]
ba_degree_counts = np.bincount(ba_degrees)
ba_x = np.arange(len(ba_degree_counts))
ba_y = ba_degree_counts / float(sum(ba_degree_counts))

# WS (Watts-Strogatz) Model
ws_graph = nx.watts_strogatz_graph(n=N, k=10, p=0.2)  # Example: 1000 nodes, each node is connected to its 10 nearest neighbors, rewiring probability of 0.2
ws_degrees = [deg for _, deg in ws_graph.degree()]
ws_degree_counts = np.bincount(ws_degrees)
ws_x = np.arange(len(ws_degree_counts))
ws_y = ws_degree_counts / float(sum(ws_degree_counts))

# Empirical Networks (Example: Assuming you have degree data stored in empirical_degrees)
empirical_degrees = [10, 15, 20, 20, 25, 25, 30, 30, 35, 40]  # Example: Degree distribution data for empirical network
empirical_degree_counts = np.bincount(empirical_degrees)
empirical_x = np.arange(len(empirical_degree_counts))
empirical_y = empirical_degree_counts / float(sum(empirical_degree_counts))

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(random_x, random_y, 'b.', label='Random')
plt.loglog(ba_x, ba_y, 'r.', label='BA Model')
plt.loglog(ws_x, ws_y, 'g.', label='WS Model')
plt.loglog(empirical_x, empirical_y, 'm.', label='Empirical')
plt.xlabel('Degree (log scale)')
plt.ylabel('Probability (log scale)')
plt.title('Degree Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
